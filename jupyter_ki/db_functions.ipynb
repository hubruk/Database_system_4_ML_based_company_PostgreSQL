```python
def do_query(query, print_out = False, commit = False):
    try:
        con = psycopg2.connect(
                    host = "127.0.0.11",
                    database="ZSBD_p1",
                    user = "inzdanych123",
                    password = "inzdanych123",
                    port = "5432")

        cur = con.cursor()

        cur.execute(query)
        if commit:
            con.commit()

        if print_out:
            row = cur.fetchone()
            while row is not None:
                print(row)
                row = cur.fetchone()

        cur.close()
        con.close()

    except (Exception, psycopg2.DatabaseError) as error:
            print(error)

    finally:
        if con is not None:
            con.close()
```

```python
def function_def(func, commit = False):
    try:
        con = psycopg2.connect(
                    host = "127.0.0.11",
                    database="ZSBD_p1",
                    user = "inzdanych123",
                    password = "inzdanych123",
                    port = "5432")

        cur = con.cursor()

        cur.execute(func)
        if commit:
            con.commit()

        cur.close()
        con.close()

    except (Exception, psycopg2.DatabaseError) as error:
            print(error)

    finally:
        if con is not None:
            con.close()
```

```python
def function_tester(function, *arguments_of_it):
    try:
        con = psycopg2.connect(
                    host = "127.0.0.11",
                    database="ZSBD_p1",
                    user = "inzdanych123",
                    password = "inzdanych123",
                    port = "5432")

        cur = con.cursor()
        
        if len(arguments_of_it) == 1:
            cur.execute("select " + function + "('" +str(arguments_of_it[0]) + "')")
        else:
            cur.execute("select " + function + str(arguments_of_it))

        row = cur.fetchone()
        while row is not None:
            print(row)
            row = cur.fetchone()

        cur.close()
        con.close()

    except (Exception, psycopg2.DatabaseError) as error:
            print(error)

    finally:
        if con is not None:
            con.close()
```

```python
function_def("""
    CREATE OR REPLACE FUNCTION ai_med.get_sum(
        a NUMERIC, 
        b NUMERIC) 
    RETURNS NUMERIC AS $$
    BEGIN
        RETURN a + b;
    END; $$

    LANGUAGE plpgsql;
    """,
    commit = True)
```

```python
function_tester("ai_med.get_sum", 10, 20)
```
